## Cleaning

```python
backup=df.copy()
```

```python
df['Price']=df['Price'].str.replace('₹','')
df['Price']=df['Price'].str.replace(',','')
df['Price']=df['Price'].str.replace(' ','')
```

```python
mask=~(df['Price']=='AskForPrice')
df=df[mask]
```

```python
df['Price'].unique()
```

```python
df['Price']=df['Price'].astype(int)
```

```python
df['Kms_driven']=df['Kms_driven'].str.replace('kms|\s+|,','',regex=True)
```

```python
df['Kms_driven']=df['Kms_driven'].astype(int)
```

```python
df['Kms_driven'].dtype
```

```python
df.info()
```

```python
df['Fuel_type'].isna().sum()
```

```python
## keep the first three words of the name
df['Name']=df['Name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df=df[df['Price']<7e6].reset_index(drop=True)
```

```python
df.drop(columns=['Label','Location'],inplace=True)
```

```python
df
```

```python
df['Fuel_type']=df['Fuel_type'].str.strip()
```

```python
# df['Owner'] = df['Owner'].str.extract(r'(\d+)')  # Keeps only the number part
```

```python
df['Owner'].isna().sum()
```

```python
mask = ~(df['Owner'].isna())
df = df[mask]
```

```python
df['Owner']=df['Owner'].str.strip()
```

```python
df.to_csv('cleaned_car.csv')
```

# Model

```python
X=df.drop(columns='Price')
Y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
df.columns
```

```python
ohe=OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[['Name','Company','Fuel_type','Owner']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Name','Company','Fuel_type','Owner']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
column_trans
```

```python
pipe.fit(xtrain,ytrain)
```

```python
ypred=pipe.predict(xtest)
```

```python
scores=[]
for i in range(1000):
  xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=.2,random_state=i)
  lr=LinearRegression()
  pipe=make_pipeline(column_trans,lr)
  pipe.fit(xtrain,ytrain)
  ypred=pipe.predict(xtest)
  scores.append(r2_score(ytest,ypred))



```

```python
import numpy as np
```

```python
scores[np.argmax(scores)]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(xtrain,ytrain)
ypred=pipe.predict(xtest)
r2_score(ytest,ypred)
```
